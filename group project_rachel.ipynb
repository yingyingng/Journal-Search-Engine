```python
import json

f = open("master.json", encoding="utf8")

publications = json.loads(f.read())

field_no = int(input ("How many fields would you like to search? "))

field = {"author", "journal", "title", "keyword"}
i = 0 
while i < field_no:
    field_type = input("Name of field (author, journal, title or keyword): ")
    if field_type in field:
        i = i + 1
        if field_type == "author":
            author_search = input ("Name of author: ") #str?
            #author_search = author_search.lowercase()
        elif field_type == "journal":
            journal_search = input ("Name of journal should contain: ")
            #journal_search = journal_search.lowercase()
        elif field_type == "title":
            title_search = input("Title of article should contain: ")
            #title_search = title_search.lowercase()
        elif field_type == "keyword":
            keyword_search = input ("Keyword: ")
            #keyword_search = keyword_search.lowercase()
    else:
        print (f"{field_type} is not a valid field")             #this counts as part of the number of fields, idk how to ignore it as a count
    #i = i + 1
    
if field_no > 1:
    and_or = {"and", "or"}
    and_or_input = input ("Type 'and' to apply all fields and 'or' for any: ")
    if and_or_input in and_or:
        if and_or_input == "and":
            i = 0
            for publication in publications:   
                if field_type == "author":
                    author_list = []
                    for author in publication["authors"]:
                        author_list.append(author["lname"])
                        author_list.append(author["fname"])
                if field_type == "journal":
                    journal_list = []
                    for journal in publication["source"]["journal"]:
                        journal_list.append(journal["source"]["journal"])
                if field_type == "title":
                    title_list = []
                    for title in publication["paper"]["title"]:
                        title_list.append(title["paper"]["title"])
                if field_type == "keyword":
                    keyword_list = []
                    for keyword in publication["paper"]["keywords"]:
                        keyword_list.append(keyword["paper"]["keywords"])
                if author_search in author_list and journal_search in journal_list and title_search in title_list and keyword_search in keyword_list:
                    print ("Paper ID:", publication ["paper_id"])
                    print (publication["paper"]["title"])
                    print ("\n")
                    i = i + 1
                
        elif and_or_input == "or":
            i = 0
            for publication in publications:
                if field_type == "author":
                    author_list = []
                    for author in publication["authors"]:
                        author_list.append(author["lname"])
                        author_list.append(author["fname"])
                if field_type == "journal":
                    journal_list = []
                    for journal in publication["source"]["journal"]:
                        journal_list.append(journal["source"]["journal"])
                if field_type == "title":
                    title_list = []
                    for title in publication["paper"]["title"]:
                        title_list.append(title["paper"]["title"])
                if field_type == "keyword":
                    keyword_list = []
                    for keyword in publication["paper"]["keywords"]:
                        keyword_list.append(keyword["paper"]["keywords"])
                if author_search in author_list or journal_search in journal_list or title_search in title_list or keyword_search in keyword_list:
                    print ("Paper ID:", publication ["paper_id"])
                    print (publication["paper"]["title"])
                    print ("\n")
                    i = i + 1 
    else:
        print (f"{and_or_input} is not a valid input")
        
elif field_no == 1:
    i = 0
    for publication in publications:   
        if field_type == "author":
            author_list = []
            for author in publication["authors"]:
                author_list.append(author["lname"])
                author_list.append(author["fname"])
                if author_search in author_list:                       
                    print ("Paper ID:", publication ["paper_id"])
                    print (publication["paper"]["title"])
                    print ("\n")
                    i = i + 1
        elif field_type == "journal":
            journal_list = []
            for journal in publication["source"]["journal"]:
                journal_list.append(journal["source"]["journal"])
                if journal_search in journal_list:                       
                    print ("Paper ID:", publication ["paper_id"])
                    print (publication["paper"]["title"])
                    print ("\n")
                    i = i + 1
        elif field_type == "title":
            title_list = []
            for title in publication["paper"]["title"]:
                title_list.append(title["paper"]["title"])
                if title_search in title_list:                       
                    print ("Paper ID:", publication ["paper_id"])
                    print (publication["paper"]["title"])
                    print ("\n")
                    i = i + 1
        elif field_type == "keyword":
            keyword_list = []
            for keyword in publication["paper"]["keywords"]:
                keyword_list.append(keyword["paper"]["keywords"])
                if keyword_search in keyword_list:                       
                    print ("Paper ID:", publication ["paper_id"])
                    print (publication["paper"]["title"])
                    print ("\n")
                    i = i + 1
```
